```python
#Task5
import seaborn as sns
df_iris = sns.load_dataset("iris")
df_titanic = sns.load_dataset("titanic")
df_tips = sns.load_dataset("tips")

iris_mean = df_iris.groupby('species').mean()
iris_std = df_iris.groupby('species').std()
print(iris_mean)
print(iris_std)

titanic_mean = df_titanic.groupby("survived")[["age", "fare"]].mean()
print(titanic_mean)

tips_mean=  df_tips[["total_bill","tip"]].mean()
tips_median=  df_tips[["total_bill","tip"]].median()
print(tips_mean)
print(tips_median)
```

```python
#Task7
import seaborn as sns
import matplotlib.pyplot as plt

df_tips = sns.load_dataset("tips")

Q1 = df_tips["total_bill"].quantile(0.25)
Q3 = df_tips["total_bill"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_tips[(df_tips["total_bill"] < lower_bound) |
(df_tips["total_bill"] > upper_bound)]
outliers

sns.boxplot(x=df_tips["total_bill"])
plt.show


```

```python
Q1 = df_tips["tip"].quantile(0.25)
Q3 = df_tips["tip"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_tips[(df_tips["tip"] < lower_bound) |
(df_tips["tip"] > upper_bound)]
outliers

sns.boxplot(x=df_tips["tip"])
plt.show
```
