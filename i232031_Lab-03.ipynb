```python
#Task#2
Paths={
    "W": {"distance": 300, "time": 25, "congestionLevel": "Medium"},
    "X": {"distance": 200, "time": 15, "congestionLevel": "Low"},
    "Y": {"distance": 150, "time": 20, "congestionLevel": "Medium"},
    "Z": {"distance": 250, "time": 10, "congestionLevel": "High"}
}
Congestions={
    'High':1,
    'Medium':2,
    'Low':3,
}
def utility(path):
    distance=path['distance']
    time=path['time']
    congestionLevel=Congestions[path['congestionLevel']]
    return-(distance*2+time*3.5+congestionLevel*1)

values={}

for name,info in Paths.items():
    values[name]=utility(info)

best=max(values,key=values.get)

for path, score in values.items():
    print("Path",path,":",score)

print("Path:",best,"with best values",values[best])
```

```python
#Task3
environment={
    "section1":"Healthy",
    "section2":"Wilted"
}

state={
    "section1":"Unknown",
    "section2":"Unknown"
}

location = "section1"

def observe(section):
    return environment[section]

def water(section):
    environment[section]="Healthy"

def move(section):
    if section=="section1":
        return "section2"
   
    if section=="section2":
        return "section1"

steps=6

for i in range(steps):
    print("Step",i+1)
    print("Location:",location)

    perception=observe(location)
    state[location]=perception

    if perception=="Wilted":
        action="Water"
        water(location)
    else:
        action="Move"
        location=move(location)

    print("Action:",action)
    print("State:",state)
    print("Environment:",environment)
```
